```python
def gpa_calculator():


    list_2unit = []
    list_3unit = []

    def grade_checker(grade):
        grade = str(grade)
        checker = 0
        while(checker == 0):
            if ((grade == 'a') or (grade == 'b')):
                checker += 1
                return grade
            elif ((grade == 'c') or (grade == 'd')):
                checker += 1
                return grade
            elif ((grade == 'e') or (grade == 'f')):
                checker += 1
                return grade
            else:
                checker = 0
                grade = str(input('Enter a valid grade (e.g A, B, C...) : '))
    
    def grade_converter(grade):
        grade = str(grade)
        grade = grade.lower()
        score = 0
        if (grade == 'a'):
            return 5
        elif (grade == 'b'):
            return 4
        elif (grade == 'c'):
            return 3
        elif (grade == 'd'):
            return 2
        elif (grade == 'e'):
            return 1
        elif (grade == 'f'):
            return 0
        else:
            print('Wrong Input!')

    gst125 = str(input("Input GST125 grade: "))
    gst125 = grade_checker(gst125)
    list_2unit.append(gst125)

    gst104 = str(input("Input GST104 grade: "))
    gst104 = grade_checker(gst104)
    list_2unit.append(gst104)

    gst105 = str(input("Input GST105 grade: "))
    gst105 = grade_checker(gst105)
    list_2unit.append(gst105)

    mth102 = str(input("Input MTH102 grade: "))
    mth102 = grade_checker(mth102)
    list_3unit.append(mth102)

    gst102 = str(input("Input GST102 grade: "))
    gst102 = grade_checker(gst102)
    list_3unit.append(gst102)

    mth103 = str(input("Input MTH103 grade: "))
    mth103 = grade_checker(mth103)
    list_3unit.append(mth103)

    phy102 = str(input("Input PHY102 grade: "))
    phy102 = grade_checker(phy102)
    list_3unit.append(phy102)

    csc102 = str(input("Input CSC102 grade: "))
    csc102 = grade_checker(csc102)
    list_3unit.append(csc102)

    phy108 = str(input("Input PHY108 grade: "))
    phy108 = grade_checker(phy108)

    total_3unit = 0
    total_2unit = 0

    for x in list_3unit:
        score = (grade_converter(x)) * 3
        total_3unit = total_3unit + score

    for x in list_2unit:
        score = (grade_converter(x)) * 2
        total_2unit = total_2unit + score

    total_1unit = grade_converter(phy108)

    cumulative = (total_3unit + total_2unit + total_1unit)

    gpa = (cumulative / 110) * 5

    print(gpa) # next you call the function if it runs with no indentation error. Otherwise, lemme know.
```
